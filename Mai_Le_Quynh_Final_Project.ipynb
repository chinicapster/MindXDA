```python
df3 = df[df['Target']!=1] # do nhóm Enrolled không có giá trị kết luận cụ thể nên việc chạy mô hình sẽ chỉ tập trung vào nhóm Graduate và Dropout
df3.info()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import sklearn.metrics as metrics
```

```python
X=df3.drop(['Target'], axis=1)
```

```python
y=df3.Target[:]
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
rf = RandomForestClassifier()
dtr = DecisionTreeClassifier()
lr = LogisticRegression()
gnb = GaussianNB()
svm = SVC(probability=True)
```

```python
rf.fit(X_train,y_train)
```

```python
dtr.fit(X_train,y_train)
```

```python
lr.fit(X_train,y_train)
```

```python
gnb.fit(X_train,y_train)
```

```python
svm.fit(X_train,y_train)
```

```python
rf_predict = rf.predict(X_test)
dtr_predict = dtr.predict(X_test)
lr_predict = lr.predict(X_test)
gnb_predict = gnb.predict(X_test)
svm_predict = svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
rf_accuracy=accuracy_score(y_test,rf_predict)
dtr_accuracy=accuracy_score(y_test,dtr_predict)
lr_accuracy=accuracy_score(y_test,lr_predict)
gnb_accuracy=accuracy_score(y_test,gnb_predict)
svm_accuracy=accuracy_score(y_test,svm_predict)
print("rf_accuracy là",rf_accuracy)
print("dtr_accuracy là",dtr_accuracy)
print("lr_accuracy là",lr_accuracy)
print("gnb_accuracy là",gnb_accuracy)
print("svm_accuracy là",svm_accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm_rf = confusion_matrix(y_test, rf_predict, labels=rf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=rf.classes_)
disp.plot(cmap=plt.cm.Reds)
disp.ax_.set_title('Confusion Matrix - Random Forest')
plt.show()
```

```python
cm_dtr = confusion_matrix(y_test, dtr_predict, labels=dtr.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_dtr, display_labels=dtr.classes_)
disp.plot(cmap=plt.cm.Reds)
disp.ax_.set_title('Confusion Matrix - Decision Tree')
plt.show()
```

```python
cm_lr = confusion_matrix(y_test, lr_predict, labels=lr.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=lr.classes_)
disp.plot(cmap=plt.cm.Reds)
disp.ax_.set_title('Confusion Matrix - Logistic Regression')
plt.show()
```

```python
cm_gnb = confusion_matrix(y_test, gnb_predict, labels=gnb.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_gnb, display_labels=gnb.classes_)
disp.plot(cmap=plt.cm.Reds)
disp.ax_.set_title('Confusion Matrix - Naive Bayes')
plt.show()
```

```python
cm_svm = confusion_matrix(y_test, svm_predict, labels=svm.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_svm, display_labels=svm.classes_)
disp.plot(cmap=plt.cm.Reds)
disp.ax_.set_title('Confusion Matrix - SVM')
plt.show()
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
dtr_accuracy=accuracy_score(y_test,dtr_predict)
lr_accuracy=accuracy_score(y_test,lr_predict)
gnb_accuracy=accuracy_score(y_test,gnb_predict)
svm_accuracy=accuracy_score(y_test,svm_predict)
print("rf_accuracy là",rf_accuracy)
print("dtr_accuracy là",dtr_accuracy)
print("lr_accuracy là",lr_accuracy)
print("gnb_accuracy là",gnb_accuracy)
print("svm_accuracy là",svm_accuracy)
```

```python
rf_accuracy=accuracy_score(y_test,rf_predict)
rf_precision=precision_score(y_test,rf_predict,average='macro')
rf_recall=recall_score(y_test,rf_predict,average='macro')
rf_f1=f1_score(y_test,rf_predict,average='macro')
print("Mô hình Random Forest")
print("Accuracy:","{:.2%}".format(rf_accuracy))
print("Precision:","{:.2%}".format(rf_precision))
print("Recall:","{:.2%}".format(rf_recall))
print("F1:","{:.2%}".format(rf_f1))
```
